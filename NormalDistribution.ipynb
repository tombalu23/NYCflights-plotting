Demonstrating properties of Central Limit Theorem

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statistics import mean,stdev
```

```
flights=pd.read_csv('/gdrive/My Drive/Data_Lit/formatted_flights.csv')

#split data
sample=flights['arr_delay'].sample(n=1000, random_state=1)
print(sample)
print(sample.mean())
```

```
samples_mean=[]
i=0

for j in range (0,10000):
  i+=1
  sample=flights['arr_delay'].sample(n=1000, random_state=i)
  samples_mean.append(sample.mean())
```

```
print(samples_mean)
print(len(samples_mean))

sm=pd.DataFrame(samples_mean)
print(flights['arr_delay'].mean()) #mean of original data
print(sm.mean())  #mean of means
```

Mean is pretty close to mean of orginal data

```
sns.distplot(sm, hist='True')
```

The above plot resembles that of a normal distribution

The observations suggest that Central Limit Theorem holds good here
